```python
import aopy
from aopy.data import db
from aopy.analysis import accllr
import os
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
import h5py
import traceback
from scipy.stats import zscore
import datetime
from tqdm.auto import tqdm
from IPython.display import display, Markdown
import pandas as pd
from tqdm.auto import tqdm
import datetime

from aopy.preproc.bmi3d import get_laser_trial_times
from aopy.data.bmi3d import tabulate_poisson_trial_times
from aopy.data.bmi3d import tabulate_poisson_trial_times
from aopy.data.bmi3d import tabulate_stim_data

data_dir = '/media/moor-data/raw'
preproc_dir = '/media/moor-data/preprocessed'
postproc_dir = '/data/postprocessed/leo'
version = 'v251029'
```

```python
elec_pos, acq_ch, elecs = aopy.data.load_chmap()
```

```python
def filter_fn(e):
    return ((len(e.record.report) > 0) and (int(e.record.report[12]) > 0) and
            (e.get_task_param('headstage_channels') == [1,256]))
```

```python
def save_df(df, postproc_dir, subject, task, version):
    filename = f'postproc_{subject}_{task}_stim_{version}.pkl'
    filepath = os.path.join(postproc_dir, subject, filename)
    pd.to_pickle(df, filepath)
    print(filepath)
```
