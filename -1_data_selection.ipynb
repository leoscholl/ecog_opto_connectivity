```python
import aopy
from aopy.data import db
import numpy as np
from tqdm.auto import tqdm
import traceback
```

```python
from connectivity_analysis import *
from connectivity_plotting import *
```

```python
preproc_dir = '/data/preprocessed'
```

```python
entries = db.lookup_sessions(subject='beignet', date=('2022-01-01', '2022-07-01'), task_name='laser only', 
                             task_desc='laser stimulation', exclude_ids=[4332, 5967, 10826])
```

```python
subjects, ids, dates = db.list_entry_details(entries)
df = aopy.data.tabulate_stim_data(preproc_dir, subjects, ids, dates, debug=False, metadata=['stimulation_site', 'drmap_chamber', 'drmap_drive_type'])
```

```python
# Which sites are available
df['stimulation_site'] = df['stimulation_site'].astype(int)
print(np.unique(df['stimulation_site']))
```

```python
# Divide by stimulation site
for stim_site, df_site in df.groupby('stimulation_site'):

    # remove any days with < 350 trials if possible
    trials = []
    days = []
    for day, df_day in tqdm(df_site.groupby('date')):
        days.append(day)
        trials.append(len(df_day))
    
    if np.any(np.array(trials) >= 350):
        for day_idx in range(len(days)):
            if trials[day_idx] < 350:
                idx = df_site.index[df_site['date'] == days[day_idx]]
                df_site.drop(idx, axis='index', inplace=True)
    
    erp_days = []
    days = []
    ids = []
    trials = []
    for day, df_day in tqdm(df_site.groupby('date')):

        df_day = df_day.reset_index(drop=True)

        if len(df_day) == 0:
            print('no data on day', day)
            continue

        try:
            # Calculate erp
            max_erp = calc_opto_resp(df_day)
            erp_days.append(max_erp)
            days.append(day)
            ids.append(np.unique(df_day['te_id']))
            trials.append(len(df_day))

        except:
            print('skipped day', day)
            traceback.print_exc()
            

    plot_spatial_drive_maps(erp_days, (1,len(erp_days)), (1,1), cmap='bwr', clim=(-3,3), elec_data=True,
                            interp=True, grid_size=(16,16), interp_method='cubic', threshold_dist=1.0, 
                            nan_color='white', label_mode=None)

    # Pick the day with the highest erp
    sort_idx = np.argmax(abs(np.sum(erp_days, axis=1)))
    
    print(f'site {stim_site}: {days[sort_idx]} (ids: {ids[sort_idx]}, {trials[sort_idx]} trials')
    plt.figure()
    im, pcm = plot_annotated_spatial_drive_map_stim(erp_days[sort_idx], stim_site, df['subject'][0], 'lm1', 0, 
                                                    grid_size=(16,16), cmap='bwr', clim=(-3,3),
                                                    colorbar=True, fontsize=9, color='k')

    plt.show()
```
